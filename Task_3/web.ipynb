```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# شغّل Chrome
driver = webdriver.Chrome()

# افتح الموقع
driver.get("https://wuzzuf.net/search/jobs?a=spbg&q=machine%20learning")

# استنى لحد ما يظهر أول عنصر من عناوين الوظائف
WebDriverWait(driver, 15).until(
    EC.presence_of_element_located((By.CLASS_NAME, "css-m604qf"))
)

# خد HTML بعد ما الصفحة تتحمل
soup = BeautifulSoup(driver.page_source, "lxml")

# كل وظيفة في بلوك class="css-1gatmva e1v1l3u10"
job_blocks = soup.find_all("div", class_="css-1gatmva e1v1l3u10")

for job in job_blocks:
    # العنوان + اللينك
    title_elem = job.find("h2", class_="css-m604qf")
    title = title_elem.text.strip() if title_elem else "N/A"
    link = "https://wuzzuf.net" + title_elem.a['href'] if title_elem and title_elem.a else "N/A"

    # اسم الشركة
    company_elem = job.find("a", class_="css-17s97q8")
    company = company_elem.text.strip() if company_elem else "N/A"

    # المكان
    location_elem = job.find("span", class_="css-5wys0k")
    location = location_elem.text.strip() if location_elem else "N/A"

    print(f"Job Title : {title}")
    print(f"Company   : {company}")
    print(f"Location  : {location}")
    print(f"Link      : {link}")
    print("-" * 50)

driver.quit()
```
